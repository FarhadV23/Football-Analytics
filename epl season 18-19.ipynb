```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
import pandas as pd

df = pd.read_csv("england-premier-league-players-2018-to-2019-stats.csv")

df
```

```python
df_GA_EPL = df[["full_name", "age","minutes_played_overall","appearances_overall","goals_overall","goals_home","goals_away","assists_overall","penalty_goals","penalty_misses","goals_involved_per_90_overall",'assists_per_90_overall','goals_per_90_overall','goals_per_90_home','goals_per_90_away','min_per_goal_overall']]

df_GA_EPL
```

top 10 Goalscorers (EPL 18/19)

```python
df_Goal = df_GA_EPL.sort_values(by="goals_overall", ascending=False).head(10)

df_Goal
```

```python
plt.scatter(
    df_Goal['full_name'],df_Goal['goals_overall'],
    color='red',
    marker='o',
)

plt.xlabel("Players")
plt.ylabel("Goal Scored")

plt.xticks(rotation=45, ha="right", fontsize=10)

plt.subplots_adjust(bottom=0.25)

plt.show()
```

Goals home and away comp (from top 10 goalscorers)

```python
players = df_Goal['full_name']
goals_home = df_Goal['goals_per_90_home']
goals_away = df_Goal['goals_per_90_away']


x = np.arange(len(players))
width = 0.4

fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(x - width/2, goals_home, width, label="Goals per 90 (Home)", color='blue')
ax.bar(x + width/2, goals_away, width, label="Goals per 90 (Away)", color='red')

ax.set_xlabel("Player")
ax.set_ylabel("Goals per 90 minutes")
ax.set_title("Comparison of Goals per 90 Minutes (Home vs Away)")
ax.set_xticks(x)
ax.set_xticklabels(players, rotation=45, ha="right") 

ax.legend()

plt.show()
```

Ronaldo VS Suarez

```python
season = ["14/15", "15/16", "16/17", "17/18", "18/19"]  # Названия сезонов
ronaldo_goals = [48, 35, 25, 26, 21]
ronaldo_as = [17, 11, 6, 5, 9]
suarez_goals = [16, 40, 28, 25, 19]
suarez_as = [16, 18, 15, 13, 8]
```

```python
x = np.arange(len(season))
width = 0.4

fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(x - width/2, ronaldo_goals, width, label="Ronaldo Goals", color='blue')
ax.bar(x - width/2, ronaldo_as, width, bottom=ronaldo_goals, label="Ronaldo Assists", color='lightblue')

ax.bar(x + width/2, suarez_goals, width, label="Suarez Goals", color='red')
ax.bar(x + width/2, suarez_as, width, bottom=suarez_goals, label="Suarez Assists", color='salmon')

ax.set_xlabel("Season")
ax.set_ylabel("Count")
ax.set_title("Goals + Assists Comparison(League) : Ronaldo vs Suarez")
ax.set_xticks(x)
ax.set_xticklabels(season)

ax.legend()

plt.show()
```
